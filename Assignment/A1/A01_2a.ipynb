```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Read and declare the class of graph data from gml files
g_karate = nx.read_gml("data/graph_karate.gml")
g_netsci = nx.read_gml("data/graph_netsci.gml")
g_facebook = nx.read_gml("data/graph_facebook.gml")
```

# 1. the average degree **< k >** and the density **$\delta$**

## Using 'Zachary karate club' Dataset.

```python
#Find the Dict degree of the network and create the List of degrees.
g_karate_degrees = g_karate.degree()
g_karate_k_list = [g_karate_degrees[i] for i in g_karate.nodes()]

# Find the properties of the network.
num_g_karate_nodes = len(g_karate.nodes())
sum_g_karate_k = np.sum(g_karate_k_list) 

avg_g_karate_k = sum_g_karate_k / num_g_karate_nodes

# not (n-1)*n, but (n-1)*n // 2 because it is an undirected graph.
max_g_karate_k = num_g_karate_nodes * (num_g_karate_nodes-1) // 2 

# Sum of K should be divided by 2 to avoid double-counting under the undirected & non-self loop condition.
# Same with nx.density(<<graph>>)
density_g_karate = (sum_g_karate_k / 2) / max_g_karate_k



# Display the result.
print('======== g_karate ========')
print(*g_karate_k_list[:10], sep=", ")
print('...')
print('# of k...', len(g_karate_k_list))
print('sum of k...',sum_g_karate_k)
print('theoritical max of total k...',max_g_karate_k)
print('---------Answer---------------')
print('Average of <k>...',avg_g_karate_k)
print('Network Density','\N{greek small letter delta}','...',density_g_karate)

# Draw the graph.
nx.draw(g_karate, with_labels = False, node_size=10000//num_g_karate_nodes)
```

# 2. Write a function **'max_degree'** that takes a network as its argument, and returns two values:
    - the id/name/label of the node with the largest degree,
    - and the value of its degree.

## define the max_degree function with an network argument and two values return.


```python
def max_degree(g):

    g_k_dict = dict(g.degree)
    g_k_label_list = list(g_k_dict)

    max_k_label = max(g_k_dict, key = g_k_dict.get)
    max_k_id = g_k_label_list.index(max_k_label)
    max_k = g_k_dict[max_k_label]

    attributes_str = 'id: ' + str(max_k_id) + '/ label: ' + str(max_k_label)

    #print ('id', max_k_id)
    #print ('label...', max_k_label)
    #print ('k_i...', max_k)

    return (attributes_str, max_k)
```

```python
g_karate_max_k_attr =max_degree(g_karate)[0]
g_karate_max_k = max_degree(g_karate)[1]

print('======== g_karate ========')
print('Node Attributes with the largest k')
print('...', g_karate_max_k_attr)
print('The largest k')
print('...',g_karate_max_k)
print('------------------------')
```
