# Assignment A01.1 Building and visualising networks
*Task: Build and visualise the following networks*

1. Star Network
2. Fully-connected
3. One-dimensional lattice with periodic boundary conditions where each node i has degree k_i = 2

Parameters : Build each network with **N = 32** nodes. Note that these networks are undirected and unweighted.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

NUM_NODES = 32
```

# 1. Star Network

```python
# Define Graph Class for Star Network.
STARNET = nx.Graph()

# Add nodes and edges composing the Network.
for i in range(NUM_NODES):
    STARNET.add_node(i)

    if i == NUM_NODES-1:
        for j in range(NUM_NODES-1):
            STARNET.add_edge(i,j)

# Print out the result and draw the network.
print ('Star Network with nodes...', len(STARNET.nodes()))
print ('Star Network Degrees (k_i)...', STARNET.degree())
nx.draw(STARNET, with_labels=True)
```

# 2. Fully-connected

```python
# Define Graph Class for Fully-connected
FULL_CONNCT = nx.Graph()

# Add nodes and edges composing the network
for i in range(NUM_NODES):
    FULL_CONNCT.add_node(i)

for i in range(NUM_NODES):
    for j in range(NUM_NODES):
        if i != j:
            FULL_CONNCT.add_edge(i,j)

# Print out the result and draw the network.
print ('Fully-Connected with nodes...', len(FULL_CONNCT.nodes()))
print ('Fully-Connected Degrees (k_i)...', FULL_CONNCT.degree())
nx.draw(FULL_CONNCT, with_labels=True)
```

# 3. One-dimensional lattice with periodic boundary conditions (k_i = 2)

```python
# Define Graph Class for One-dimensional lattice.
ONE_D_LATT_w_PDC = nx.Graph()

# Add nodes and edges composing the Network.
for i in range(NUM_NODES):
    ONE_D_LATT_w_PDC.add_node(i)
    if i < NUM_NODES-1:
        ONE_D_LATT_w_PDC.add_edge(i,i+1)
    else:
        ONE_D_LATT_w_PDC.add_edge(i,0)

# Print out the result and draw the network.
print ('One-dim lattice with PDC with nodes...', len(ONE_D_LATT_w_PDC.nodes()))
print ('One-dim lattice with PDC Degrees (k_i)...', ONE_D_LATT_w_PDC.degree())
pos = nx.circular_layout(ONE_D_LATT_w_PDC)
nx.draw(ONE_D_LATT_w_PDC, with_labels = True, pos = pos)
```
