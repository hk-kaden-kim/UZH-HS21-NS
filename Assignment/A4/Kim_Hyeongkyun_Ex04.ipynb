# A04.1 Community detection

*Task: For the given networks find the communities using (a) the greedy modularity maximization by Clauset Newman and Moore and (b) the label propagation algorithm. Assign to each community a color and draw the resulting
graph.*

*Task: Randomise each network and compare the number of communities obtained before and after randomisation.*

```python
#colors = [ 'blue', 'orange', 'green', 'red', 'purple', 'brown', 
#'pink', 'gray', 'olive', 'cyan', 'black', 'brown', 'chocolate', 
#'gold', 'orchid', 'khaki' ]

def random_color():
    return "#{}{}{}{}{}{}".format(*(random.choice("0123456789abcdef") for _ in range(6)))

colors = [random_color() for _ in range(50)]
```

## [ANSWER] Task. Generate ER graphs with 𝑁 = 100 and 𝑝 ∈ [0, 1]

1. Plot the probability that a node belongs to the largest
connected component 𝑁𝐺/𝑁 as a function of 𝑝 and mark
with a vertical line the critical probability 𝑝𝑐 = 1/𝑁

2. Plot the average clustering ⟨𝐶⟩ as a function of 𝑝 and give
an interpretation of the result

**[Self Interpretation]**

Erdos-Renyi Random networks are composing the network based on the probability of an existing edge between two nodes. And this parameter is independent of any other edges. In hence of this characteristic of Erdos-Renyi Random network, the average of clustering coefficient, which means the likelihood between two neighbors are linked with each other, is measured closely the same with the p-value of Erdos-Renyi Random network when we repeat and average the *<C(p)>* data much more.


```python
N = 100
p_list = np.arange(0,1,0.001)
p_list = np.append(p_list, [1])

T = [100,10]
prob_NG_arr = np.zeros((T[0], len(p_list)))
avg_C_arr = np.zeros((T[1], len(p_list)))

for i in range(T[0]):
    j=0
    for p in p_list:
        
        G = nx.erdos_renyi_graph(N, p)

        # Find the Giant Component
        Comp = [c for c in nx.connected_components(G)]
        maxComp = max(Comp, key = len)
        NG = len(maxComp)
        # Calculate the probability of NG
        prob_NG = NG/N
        prob_NG_arr[i][j] = prob_NG

        j += 1

for i in range(T[1]):
    j=0
    for p in p_list:
        
        G = nx.erdos_renyi_graph(N, p)

        #Calculate the average of clustering coefficients 
        avg_C = nx.average_clustering(G)
        avg_C_arr[i][j] = avg_C
        
        j += 1

prob_NG_mean = np.average(prob_NG_arr, axis = 0)
avg_C_mean = np.average(avg_C_arr, axis = 0)
```

```python
fig1 = plt.figure(figsize = (10,10))
plt.subplots_adjust(wspace = 0.2, hspace = 0.5)

axs1 = fig1.add_subplot(211)
axs1.plot(p_list, prob_NG_mean)
axs1.set_title('1. NG(p)/N')
axs1.set_xlabel('p (log)')
axs1.set_ylabel('NG/N')
axs1.set_xscale('log')
axs1.axvline(x=1/N, color = 'red')

axs2 = fig1.add_subplot(212)
axs2.plot(p_list, avg_C_mean)
axs2.set_title('2. <C(p)>')
axs2.set_xlabel('p')
axs2.set_ylabel('<C>')
#axs2.set_xscale('log')
```

## [ANSWER] Task. WS small-world networks (𝑁 = 100, 2𝜅 = 10)

```python
N = 100
k = 10
knn_cord = k/2

p_list = np.arange(0,1,0.001)
p_list = np.append(p_list, [1])

T = 10
avg_C_anlyt_arr = []

WS_net_ref = nx.watts_strogatz_graph(N, k, 0)
avg_C_ref = nx.average_clustering(WS_net_ref)
D_ref = nx.average_shortest_path_length(WS_net_ref)

avg_C_arr = np.zeros((T, len(p_list)))
D_arr = np.zeros((T, len(p_list)))

for i in range(T):
    j=0
    for p in p_list:
        WS_net = nx.watts_strogatz_graph(N, k, p)

        avg_C = nx.average_clustering(WS_net)
        D = nx.average_shortest_path_length(WS_net)
        if (i==0):
            #Check the analytical result of the average clustering.
            avg_C_anlyt = (3*(knn_cord - 1)*(1-p)**3)/(2*(2*knn_cord-1))/avg_C_ref
            avg_C_anlyt_arr.append(avg_C_anlyt)
        
        avg_C_arr[i][j] = avg_C/avg_C_ref
        D_arr[i][j] = D/D_ref

        j += 1

avg_C_arr_T100 = np.average(avg_C_arr, axis = 0)
D_arr_T100 = np.average(D_arr, axis = 0)
```
